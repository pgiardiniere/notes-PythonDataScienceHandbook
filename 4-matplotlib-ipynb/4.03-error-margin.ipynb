This section regards visualizing error-bars on graphs in mpl

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
```

First, make a sin wave, adding some randomness to the plot.

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.k')
```

In the above graph, the keyword argument 'fmt' specifies the format code. It shares shorthand with plt.plot() seen before.

Tossing in some additional kwargs shows off flexibility of plt.errorbar().

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black', 
             ecolor='lightgray', elinewidth=3, capsize=0)
```
