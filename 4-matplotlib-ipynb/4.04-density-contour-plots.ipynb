We can use mpl to plot 3 dimensional funcs to a 2d plane using contour/density graphs.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
# z = f(x, y)
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
# np.meshgrid() forms 2d grid from arrays x,y
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# Black/White Contour Plot: Negative vals dashed lines, positive vals solid lines
plt.contour(X, Y, Z, colors='black')
```

```python
# Colored Contour Plot:
plt.contour(X, Y, Z, 20, cmap = 'RdGy')
```

```python
# Filled Colored Contour Plot:
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

```python
# Smoothed, Filled, Colored Contour Plot:
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image');
```
