## Three-Dimensional Plots

Matplotlib has a seperate module which allows plotting on 3 axes (x, y, z).

Within the mpl_toolkits package, you'll need to import the mplot3d module.

* Note: Will be using an interactive notebook for this demonstration.

* May have only one plot in interactive mode at a time.

```python
%matplotlib notebook
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
# To plot 3-dimensional axes, pass kwarg projection='3d'
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

^^ Note that we're only getting the first octant of the 3 dimensional plane

i.e. we have:

x | 0 <= x <= a

y | 0 <= y <= b

z | 0 <= z <= c



for a, b, c | some scalars > 0.

### Points and Lines in R3

A point in R^3 is a 3-tuple, (x, y, z).

A plot of a set of such points would be a Scatterplot.

A line in R^3 can be defined as some z | z = f(x, y)

A plot of some such function z would result in a line.

```python
ax = plt.axes(projection='3d')

# Three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Three-dimensional points, scattered about the 3D line.
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

### Three-dimensional Contour Plots

Contour plots which would give pause in 2-dimensional plots will be much more immediate in a 3d plot.

Use ax.contour3D() in much the same way as ax.contour. I.e. first get your two-dimensional grids of x,y. Then define Z | some function on those grids.

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the relationship
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
# Especially when in non-interactive mode, helpful to specify a viewing angle.

# elevation denotes num degrees above the xy plane
# azimuth denotes num degrees rotation counter-clockwise about the z-axis.

# These values display in the lower-right corner of the interactive plot as you drag it around.

ax.view_init(60, 35)    # (elev, azim)
fig
```

### Wireframes and Surface Plots

Wireframes are of a similar kin to contour plots, see example below.

Surface Plots are just Wireframes where the faces are filled polygons. Assigning a colormap to polygons assists in visualizing surface.

```python
# Wireframe
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
ax.set_title('wireframe');
```

```python
# Surface Plot
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
```

```python
# Slice of 3D surface plot
r = np.linspace(0, 6, 20)
theta = np.linspace(-0.9 * np.pi, 0.8 * np.pi, 40)
r, theta = np.meshgrid(r, theta)

X = r * np.sin(theta)
Y = r * np.cos(theta)
Z = f(X, Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none');
```

### Surface Triangulations

Useful when, for example, wanting to create a solid-filled graph from a scattered set of points.

```python
# Random draws
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x, y)
```

```python
# Scatterplot visualization
ax = plt.axes(projection='3d')
ax.scatter(x, y, z, c=z, cmap='viridis', linewidth=0.5);
```

```python
# Surface Triangulation visual
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, cmap='viridis', edgecolor='none')
```
