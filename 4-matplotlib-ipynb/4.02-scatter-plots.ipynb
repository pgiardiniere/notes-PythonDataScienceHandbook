```python
# notebook setup
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, 'o', color='black')
```

Third argument represents type of symbol displayed on plot.

Below are common shorthand symbols and their results:

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

```python
# Character codes can be used together with line/color codes.
plt.plot(x, y, '-ok')
```

```python
# additional keyword arguments to plt.plot(): (more lines/markers)
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```

### Scatter plots with plt.scatter()
<hr style="border:1px solid gray"> </hr>

plt.scatter() is more powerful than the simple plt.plot().

```python
plt.scatter(x, y, marker='o')
```

Main difference being plt.scatter() is granular down to level of each individual point.

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```

So in addition to basic scatterplots, we can now use size and color of points to convey further dimensions of the data within the visualization

In the following example, we visualize 4 dimensions of data on a 2D plot.

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

Clear tradeoff between plt.scatter() and plt.plot() is granularity and speed.

i.e. for large n, plt.scatter() will perform noticeably slower than plt.plot() due to the fact that it tracks each individual point
