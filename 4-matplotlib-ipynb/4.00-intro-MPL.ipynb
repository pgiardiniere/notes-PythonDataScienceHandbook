State-based vs Object-based approaches to matplotlib

State-Based: More like Matlab. See the following example:

```python
plt.figure()            # create plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1)    # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
plt.gcf()
```

```python
plt.gca()
```

Object Based: More pythonic

```python
# first, create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.sublplots(2)

# Call plot() method on appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```
